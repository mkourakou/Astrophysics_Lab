# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import pickle
import itertools
from astropy import units as  u
from astropy.cosmology import FlatLambdaCDM
cosmo=FlatLambdaCDM(H0=70, Om0=0.3)

from scipy import constants as cst
from scipy import integrate
```

# Functions

```python
def Flambda_to_Fnu(flux_lambda, wl_angst): 
    
    """
    This function transforms flux F_\lambda to F_\nu
    input:
    -flux in units of (some luminosity units) / Angstrom
    -wavelength (spectral axis) in Angstrom
    output:
    -flux in units of (some luminosity units) / Hz
    """
    
    Angstrom = 1e10

    flux_nu = flux_lambda * (wl_angst**2. ) / (cst.c * Angstrom) 
    return flux_nu

def SolarLumi_to_ergs(intensity): 
    
    """
    This function transforms L_\odot to erg/s
    input:
    -flux in units of SolarLumi/(some spectral unit) or luminosity in units of SolarLumi
    output:
    -flux in units of erg/s/(some spectral unit) or luminosity in units of erg/s
    """
     
    return intenisty*(3.846e33)


def Calzetti(gal_nu, gal_Fnu,GAebv):

    """
    This function computes the effect of reddening in the galaxy template (Calzetti law)
    input:
    - frequencies in Hz
    - Fluxes in Fnu
    - the reddening value E(B-V)_gal
    output:
    - frequencies in Hz
    - Fluxes in Fnu Reddened 
    """
    RV = 4.05 #obscuration at V       

    gal_lambda_m = cst.c / gal_nu * 1e6   #spectral axis: from Hz ----> microns 
    wl = gal_lambda_m[::-1]       #invert for lambda (ascending order of wavelength)
    k = np.zeros(len(wl))      #απορροφηση

    w0 = tuple([wl <= 0.12])  #lower limit for Calzetti law
    w1 = tuple([wl < 0.63])
    w2 = tuple([wl >= 0.63])

    x1 = np.argmin(np.abs(wl - 0.12))
    x2 = np.argmin(np.abs(wl - 0.125))
    
    #-----  equation 4 from Calzetti 2000 -----#

    k[w2] = 2.659 * (-1.857 + 1.040 /wl[w2])+RV
    k[w1] = 2.659 * (-2.156 + (1.509/wl[w1]) - (0.198/wl[w1]**2) + (0.011/wl[w1]**3))+RV
    if (wl[x1] - wl[x2]) != 0:  # avoid division by zero
        k[w0] = k[x1] + ((wl[w0] - 0.12) * (k[x1] - k[x2]) / (wl[x1] - wl[x2])) +RV
    else:
        k[w0] = 0
    
    #-----  equation 2 from Calzetti 2000 -----#

    gal_k= k[::-1] #invert for nus
    gal_Fnu_red = gal_Fnu * 10**(-0.4 * gal_k * GAebv)
    
    return gal_nu, gal_Fnu_red

```

# Load galaxy data

```python
route_data = 'data/'
filename = 'NGC5055_UV-MIR_NED.txt'

#-------- NGC5055 (spiral LINER)-----#
colnames=['Wave', 'F_lambda', 'drop','Freq', 'F_nu']   # Angstrom , erg/cm^2/s/A, Hz, W/m^2/Hz 
df = pd.read_csv(route_data+filename, skiprows=3, delim_whitespace=True, names=colnames, header=None)
df.drop('drop', axis=1, inplace=True)
df = df[df['F_lambda']>0]

redshift =  0.001678
D = cosmo.luminosity_distance(redshift)/u.Mpc*u.Mpc.to(u.cm) #in cm

df['L_lambda'] = 4*np.pi*D**2*df['F_lambda']  #erg/s/A
df['L_nu'] = Flambda_to_Fnu(df['L_lambda'].to_numpy(), df['Wave'].to_numpy())   #erg/s/Hz
```

# Load templates and create grid

```python
%%time
route_temps = './templates/'

# ------------------ Stellar population -----------------#

raw_wavelengths = np.loadtxt(route_temps+"BPASSv22.dat", usecols=0)
raw_grid = np.loadtxt(route_temps+"BPASSv22.dat")[:,1:]
ages = np.arange(2, 53)
ages = 10**(5+0.1*(ages-2))

wavelengths = df['Wave'].to_numpy().copy() #in Angstroms
grid = np.zeros((wavelengths.shape[0], raw_grid.shape[1])) #for each wvl corrsponds a 51-long array

for i in range(raw_grid.shape[1]):
    grid[:,i] = np.interp(wavelengths, raw_wavelengths, raw_grid[:,i])

grid /= 10**6   #normalise to 1 solar mass
solar_luminosity = 3.827e33  #solar lumi to erg/s
grid *= solar_luminosity  # transform solar lumi ----> erg/s

```

# 01. Plot galaxy data with an uninformed guess

We will select a template and scale it with mass $M_\star = 10^{10}\,M_\odot$ (this is an uninformed guess).  
Then, plot the data with this guess.

Note: Measurement uncertainties are not provided in our data.



```python
# -----  Starting guess (parametres: mass/ age index) --------#
mass = 1e10  # in Solar masses
templ_idx = 42 # a number between 0 and 51

#---------- They correspond to this template/ model  ---------#
temp_nus =  (cst.c*1e10) /wavelengths
temp_Fnus = Flambda_to_Fnu(grid[:,templ_idx], wavelengths)
# make model 
model_freq0 = temp_nus
model_lumi0 = mass*temp_Fnus

# ----------  Chi-sq of our guess ---------------#
errs = 1e25*np.ones(len(df['Freq']))  #arbitrarily 10^25 (simply for chi2 use)
diffs = df['L_nu'] - model_lumi0
chisq = np.sum(diffs**2/errs**2)
print("Chi-squared value for this uninformed guess: \n", chisq)
```

```python
fig = plt.figure(figsize=(10,7))

ax = fig.add_subplot(111)
ax.scatter(df['Freq'],df['L_nu'], marker = '.',color = 'darkviolet',label = "NGC5055 (spiral)"  )
ax.scatter(model_freq0, model_lumi0, marker = '.',color = 'darkorange',label = "First guess Stellar component"  ) 
ax.grid(alpha=0.4)
ylims = [1e15, 1e31]
ax.set_ylim(ylims)
ax.set_yscale('log')
xlims = [4e12, 5e16]
ax.set_xlim(xlims)
ax.set_xscale('log')
ax.set_ylabel(r"$ L_{\nu}$ (erg/s/Hz)", fontsize = 16)
ax.set_xlabel(r"$\nu_{rest}$ (Hz)", fontsize = 16)
ax.legend(loc="lower left", fontsize = 12)

plt.show()
```

# 02. $\chi^2$ minimisation για να βρούμε το καλύτερο template με το καλύτερο scaling μάζας που αναπαριστά με έναν αστρικό πληθυσμό τον γαλαξία μας. 

```python
%%time


#--------- Fit stellar light ---------#

fit_window = [4e14, 3e16]   # V to FUV
lolim = fit_window[0]
hilim = fit_window[1]
maskfit = np.logical_and(df['Freq']>lolim, df['Freq']<hilim)
df_fit = df[maskfit].copy()

errs = 1e25*np.ones(len(df_fit['Freq'])) #arbitrary

masses = 10**np.arange(9, 13, 0.05)  # μεταξύ 10^9 και 10^13 με βήμα 0.05
templ_indices = np.arange(51)      # μεταξύ index 0 και 50

best_chisq = np.inf
best_point = [0.,0.]

n_masses = masses.shape[0]
n_temps = templ_indices.shape[0]
n_models = n_masses*n_temps 

for i in range(n_masses):
    for j in range(n_temps):    
        model_no = j + i*n_temps
            
        if not model_no % 100:
            print("Tried", model_no, "/", n_models, "models.")
            
        mass = masses[i]
        templ_idx = templ_indices[j]
        
        temp_nus =  (cst.c*1e10) /wavelengths
        temp_Fnus = Flambda_to_Fnu(grid[:,templ_idx], wavelengths)
        # make model 
        model_freq = temp_nus[temp_nus>lolim]
        model_lumi = mass*temp_Fnus[temp_nus>lolim]
                    
        diffs = df_fit['L_nu'] - model_lumi

        chisq = np.sum(diffs**2/errs**2)
            
        if chisq < best_chisq:
                best_chisq = chisq
                best_point = [i,j]
                
print("\n best chi-squared value:", best_chisq)
print("\n best parameters: \n Stellar Mass = 10^{:.2f} Solar Masses".format(np.log10(masses[best_point[0]])),  
      " \n Age = {:.3f} Myrs".format(ages[best_point[1]]/1e6),"\n" )
```

# Plot data with best fit

```python
print("\n best parameters: \n Stellar Mass = 10^{:.2f} Solar Masses".format(np.log10(masses[best_point[0]])),  
      " \n Age = {:.3f} Myrs".format(ages[best_point[1]]/1e6),"\n" )
```

```python
temp_nus =  (cst.c*1e10) /wavelengths
temp_Fnus = Flambda_to_Fnu(grid[:,best_point[1]], wavelengths)
# make model 
BestFit_freq = temp_nus
BestFit_lumi = masses[best_point[0]] * temp_Fnus
 
fig = plt.figure(figsize=(10,7))

ax = fig.add_subplot(111)
ax.scatter(df['Freq'],df['L_nu'], marker = '.',color = 'darkviolet',label = "NGC5055 (spiral)"  )
ax.scatter(BestFit_freq, BestFit_lumi, marker = '.',color = 'tomato', label = r"Best $\chi^2$ fit"  ) 
ax.grid(alpha=0.4)
ylims = [1e15, 1e31]
ax.set_ylim(ylims)
ax.set_yscale('log')
xlims = [4e12, 5e16]
ax.set_xlim(xlims)
ax.set_xscale('log')
ax.set_ylabel(r"$ L_{\nu}$ (erg/s/Hz)", fontsize = 16)
ax.set_xlabel(r"$\nu_{rest}$ (Hz)", fontsize = 16)
ax.legend(loc="lower left", fontsize = 12)
ax.text(1e13, 1e22, r"Stellar Mass estimate: $10$^{:.1f}$\; M_\odot$".format(np.log10(masses[best_point[0]])), fontsize = 18, color ="indigo")
ax.text(1e13, 1e20, r"Stellar Population Age estimate: {:.3f} Myrs".format(ages[best_point[1]]/1e6), fontsize = 18, color ="indigo")

plt.show()
```
